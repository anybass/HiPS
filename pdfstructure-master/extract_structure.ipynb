### Creating a Pipeline to extract the hierarchy for all PDFs using pdfstructure

```python
from pathlib import Path
from pdfstructure.hierarchy.parser import HierarchyParser
from pdfstructure.source import FileSource
from pdfstructure.hierarchy.traversal import traverse_in_order, traverse_level_order,   traverse_inorder_sections_with_content, get_document_depth

import traceback
import os
import shutil
import tqdm.notebook as tq

parser = HierarchyParser() 
for f in tq.tqdm(sorted(os.listdir("../data/PDFs/"))):
    if f.endswith(".pdf"): 
        try:
            source= FileSource(str("../data/PDFs/"+f)) 
            # analyse document and parse as nested data structure
            document = parser.parse_pdf(source)
            elements_flat_inorder_sections_with_content = [element for element in traverse_inorder_sections_with_content(document)]
            with open("./hierarchies/"+f[:-4]+".txt", "w") as fl:
                fl.write(str(elements_flat_inorder_sections_with_content).replace("\\n", " "))
            print("Sucess:")
            print(f)
        except Exception:
            print("Failed:")
            print(f)
            print(traceback.format_exc())
            
            continue
```
